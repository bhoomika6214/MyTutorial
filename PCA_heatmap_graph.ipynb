```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets  import load_breast_cancer
df = pd.read_csv('//content//data.csv')
df.head()
```

```python
data = load_breast_cancer()
data
breast = load_breast_cancer()
breast_dataset = breast.data
breast_dataset.shape
```

```python
breast_dataset = breast.data
breast_dataset.shape
```

```python
breast_lables = breast.target
breast_lables.shape
```

```python
lables = np.reshape(breast_lables,(569,1))
lables.shape
```

```python
breast_dataset = pd.DataFrame(breast_dataset)
breast_dataset.head()
```

```python
features = breast.feature_names
features
```

```python
breast_dataset.columns =features
breast_dataset.head()
```

```python
breast_dataset["lables"]=lables
breast_dataset.head()
```

```python
breast_dataset['lables'].replace(0,'Benign',inplace=True)
breast_dataset['lables'].replace(1,'Malignant',inplace=True)
breast_dataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:,features].values
x = StandardScaler().fit_transform(x)
print(x.shape)
```

```python
np.mean(x),np.std(x)
```

```python
principalComponents_breast = pd.DataFrame(x)
principalComponents_breast.head()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(15,15))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['lables'] == target
    plt.scatter(principalComponents_breast.loc[indicesToKeep,0],principalComponents_breast.loc[indicesToKeep,1],c = color,s= 20)
plt.legend(targets,prop={'size': 15})
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an undirected graph
G = nx.Graph()

# Add nodes and undirected edges (A -- B)
G.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'A')])

# Draw the undirected graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_size=2000, node_color="lightgreen")
plt.title("Undirected Graph")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

G.add_nodes_from(["A", "B", "C","D"])

# Add nodes and directed edges (A -> B)
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'),('C', 'D')])

# Draw the directed graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_size=2000,edge_color = 'gray',font_size=14, node_color="skyblue")
plt.title("Directed Graph")
plt.show()
```

```python
# prompt: give me code for weighted graph

import networkx as nx
import matplotlib.pyplot as plt

# Create a weighted graph
G = nx.DiGraph()

G.add_nodes_from(["A", "B", "C","D"])

edges = [("A", "B", 5), ("A", "C", 3), ("B", "C", 2), ("B", "D", 7), ("C", "D", 4)]
G.add_weighted_edges_from(edges)

edge_labels = {(u, v): w for u, v, w in edges}
edge_widths = [w / 2 for _, _, w in edges]

pos = nx.spring_layout(G)
plt.figure(figsize=(6, 5))
nx.draw(G, pos, with_labels=True, node_color="skyblue",edge_color='gray', node_size=700, font_size=14,width=edge_widths,arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels,font_size=12)
plt.title("Weighted Graph")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an undirected weighted graph
G = nx.Graph()

G.add_nodes_from(["A", "B", "C", "D"])

# Define edges with weights
edges = [("A", "B", 5), ("A", "C", 3), ("B", "C", 2), ("B", "D", 7), ("C", "D", 4)]
G.add_weighted_edges_from(edges)

# Create labels for the edges with their weights
edge_labels = {(u, v): w for u, v, w in edges}
edge_widths = [w / 2 for _, _, w in edges]

# Set the layout for the graph
pos = nx.spring_layout(G)

# Draw the graph
plt.figure(figsize=(6, 5))
nx.draw(G, pos, with_labels=True, node_color="skyblue", edge_color='gray', node_size=700, font_size=14, width=edge_widths, alpha=0.7)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

# Display the graph
plt.title("Undirected Weighted Graph")
plt.show()
```
