```python
import pandas as pd

# Create a list of student data with Branch and Section
data = {
    "Name": [f"Student {i}" for i in range(1, 61)],
    "Roll No": [i for i in range(101, 161)],
    "Age": [20 + (i % 5) for i in range(60)],  # Random ages between 20 and 24
    "Gender": ["Male" if i % 2 == 0 else "Female" for i in range(60)],
    "Place": ["City" + str(i % 10) for i in range(60)],
    "Date of Birth": [f"200{i % 10}-0{(i % 12) + 1}-0{i % 28 + 1}" for i in range(60)],  # Sample DOB
    "Branch": ["Computer Science" if i % 2 == 0 else "Mechanical" for i in range(60)],
    "Section": [f"Section {chr(65 + (i % 5))}" for i in range(60)]  # Sections A, B, C, D, E
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
file_path = "student_data_with_branch_section.csv"
df.to_csv(file_path, index=False)

print(f"CSV file saved as {file_path}")
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("//content//tested.csv")
df.head(10)
```

```python
cols=["Name","Ticket","Cabin"]
df=df.drop(cols, axis=1)
df.info()
```

```python
dummies = []
cols = ["Pclass", "Sex", "Embarked"]
for col in cols:
    dummies.append(pd.get_dummies(df[col]))
```

```python
titanic_dummies = pd.concat(dummies, axis=1)
df = pd.concat((df,titanic_dummies), axis=1)
df.head()
```

```python
df = df.drop(['Pclass','Sex','Embarked'], axis=1)
```

```python
df['Age'] = df['Age'].interpolate()
print(df)
```
